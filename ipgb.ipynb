# Problem 


Importing packages

```python
import gurobipy as gp
from gurobipy import GRB
from math import sqrt
```

Store √2 in a var

```python
s2 = sqrt(2)
```

Make a model named IP

```python
m = gp.Model("IP")
```

Make integer variables x1 and x2

```python
x1 = m.addVar(vtype=GRB.INTEGER, name="x1")
x2 = m.addVar(vtype=GRB.INTEGER, name="x2")
```

Add constraint 1 ≤ x2 ≤ √2x1
such constraint can't be expressed in one in python so break it in two constraints

```python
m.addConstr(1<=x2, "c0")
m.addConstr(x2<=s2*x1, "c1")
```

Add Objective min √2x1 − x2

```python

m.setObjective(s2*x1-x2, GRB.MINIMIZE)
```

Set timelimit

```python
m.Params.TimeLimit = 50
```

Solve the model

```python
m.optimize()
```

Results

```python
for v in m.getVars():
    print(f'{v.Varname}: {v.X:g}')
print(f'Obj: {m.ObjVal:g}')

if m.status == GRB.OPTIMAL or m.status == GRB.TIME_LIMIT:
    print(f'Total solve time: {m.Runtime:.4f} seconds')
```

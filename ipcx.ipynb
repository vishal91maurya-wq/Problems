# Problem 


Importing packages

```python
from docplex.mp.model import Model
from math import sqrt
```

Store √2 in a var

```python
s2 = sqrt(2)
```

Make a model named IP

```python
m = Model(name='IP')
```

Make integer variables x1 and x2

```python
x1 = m.integer_var(name='x1')
x2 = m.integer_var(name='x2')
```

Add Objective min √2x1 − x2

```python

m.minimize(s2*x1-x2)
```

Add constraint 1 ≤ x2 ≤ √2x1
such constraint can't be expressed in one in python so break it in two constraints

```python
m.add_constraint(1<=x2, ctname='c0')
m.add_constraint(x2<=s2*x1, ctname="c1")
```

Set timelimit

```python
m.parameters.timelimit = 50
```

Solve the model

```python
solution = m.solve()
```

Results

```python
if solution:
    print(f'Optimization Status: {m.get_solve_status()}')
    print(f'Optimal value: {solution.objective_value}')
    print(f'x1: {solution.get_value(x1)}')
    print(f'x2: {solution.get_value(x2)}')
    details = m.solve_details
    print(f'Total solve time: {details.time:.4f} seconds')
    print(f'Relative Gap: {details.mip_relative_gap:.4f}')
else:
    print("No solution exists")
```
